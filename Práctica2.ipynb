```python
# Obtener información básica del DataFrame
data = pd.read_csv ( 'MailCustomer.csv' , index_col = 'CustomerKey' , sep = ';' , encoding= " ISO -8859 -1 ")

data_info = data.info()

data_head = data.head()

data_tail = data.tail()

data_nulls = data.isnull().sum().sort_values(ascending=False)



data_info, data_head, data_tail, data_nulls
```

```python
# Recorremos las fechas, con año mes y día
for i in data['BirthDate']:
  timeData = pd.to_datetime(i)
  print("Mes: " , timeData.month)
  print("Dia: " , timeData.day)
  print("Ano: " , timeData.year)
```

```python
# Vemos las fechas incorrectas
incorrect_birthdates = data[(data['BirthDate'] > pd.Timestamp('2024-01-01')) |
                            (data['BirthDate'] < pd.Timestamp('1900-01-01'))]
print(incorrect_birthdates)
```

```python
# Corregimos fechas
data.loc[data['BirthDate'] > pd.Timestamp('2024-01-01'), 'BirthDate'] = \
    data.loc[data['BirthDate'] > pd.Timestamp('2024-01-01'), 'BirthDate'] - pd.DateOffset(years=100)
```

```python
# Validamos
print(data[(data['BirthDate'] > pd.Timestamp('2024-01-01')) |
           (data['BirthDate'] < pd.Timestamp('1900-01-01'))])
```

```python
# Para la edad
from datetime import datetime

# Calcular la edad basada en la columna BirthDate
current_date = datetime.now()
data['Age'] = data['BirthDate'].apply(lambda x: current_date.year - x.year - ((current_date.month, current_date.day) < (x.month, x.day)))

# Ajustar si aún no ha ocurrido el cumpleaños
data['Age'] -= data['BirthDate'].apply(lambda x: (x.month, x.day) > (current_date.month, current_date.day))

# Mostrar el DataFrame resultante
print(data)
```

```python
#Comprobar las edades junto la fecha de nacimiento recorriendo con un for
for i in data['BirthDate']:
  timeData = pd.to_datetime(i)
  print("Edad: " , current_date.year - timeData.year - ((current_date.month, current_date.day) < (timeData.month, timeData.day)))
  print("Fecha de nacimiento: " , timeData)
  print("--------------------------------------------------------------------------------")
```

```python
# Identificar valores nulos
missing_values = data.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print("Columnas con valores faltantes:\n", missing_columns)

# Colocar none en las columnas con valores faltantes
for column in missing_columns.index:
    data[column].fillna("None", inplace=True)
```

```python
# Verificar que no hay valores faltantes
print(data.isnull().sum().sort_values(ascending=False))
```

```python
# Analizar datos duplicados exactos
duplicate_rows = data[data.duplicated()]
print("Filas duplicadas:\n", duplicate_rows)
```
